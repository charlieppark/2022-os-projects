# Scheduling Criterions for Different Scheduling Algorithms

```python
import numpy as np
import matplotlib.pyplot as plt


width = 0.15
x = np.arange(3)
fig, ax = plt.subplots(figsize=(10,6))
criterions = ['Throughput', 'Average Waiting Time', 'Average Turnaround Time']

sjf     = [0.235 / 0.2, 9.750 / 50, 5.500 / 45]
fcfs    = [0.159 / 0.2, 58.331 / 50, 52.034 / 45]
rr      = [0.172 / 0.2, 52.897 / 50, 47.086 / 45]
colors = ['#212121', '#323232', '#0D7377', '#3BACB6', '#82DBD8', '#B3E8E5']

plt.title("Normalized Scheduling Criterions for Different Scheduling Algoithms")

plt.bar(x, fcfs, width, color=colors[0], label="FCFS")
plt.bar(x + width, rr, width, color=colors[2], label="RR")
plt.bar(x + width * 2, sjf, width, color=colors[4], label="SJF")
plt.xticks(np.arange(0, 3), labels=criterions)

plt.grid(True, axis='y')
plt.ylabel('Time Tick')
plt.xlabel('Criterions')
plt.legend()
plt.show()
```
