# Hardware Access Time for Different Replacement Algorithms

## Case 1. Memory Access Pattern - Uniform

```python
import numpy as np
import matplotlib.pyplot as plt


width = 0.1
x = np.arange(3)
fig, ax = plt.subplots(figsize=(15,6))
hardware = np.array(['Disk Access Time', 'Memory Access Time', 'Overall Access Time'])

disk    = np.array([67724912040, 88086859800, 83847792360, 79861145760, 74019406320, 89609398200, 88619748240])
disk    = disk / (disk.max() + disk.min())

memory  = np.array([31754300, 32259500, 32156700, 32057200, 31911400, 32300500, 32275800])
memory  = memory / (memory.max() + memory.min())

overall = disk + memory
overall = overall / (overall.max() + overall.min())

random  = np.array([disk[0], memory[0], overall[0]])
fifo    = np.array([disk[1], memory[1], overall[1]])
lru     = np.array([disk[2], memory[2], overall[2]])
lfu     = np.array([disk[3], memory[3], overall[3]])
mfu     = np.array([disk[4], memory[4], overall[4]])
sca     = np.array([disk[5], memory[5], overall[5]])
esca    = np.array([disk[6], memory[6], overall[6]])


colors = ['#212121', '#323232', '#0D7377', '#14FFEC', '#71C9CE', '#A6E3E9', '#CBF1F5']

plt.title("Hardware Access Time for Different Replacement Algorithms - Uniform")

plt.bar(x, random, width, alpha=1, color=colors[0], label="Random")
plt.bar(x + width, fifo, width, alpha=1, color=colors[1], label="FIFO")
plt.bar(x + width * 2, lru, width, alpha=1, color=colors[2], label="LRU")
plt.bar(x + width * 3, lfu, width, alpha=1, color=colors[3], label="LFU")
plt.bar(x + width * 4, mfu, width, alpha=1, color=colors[4], label="MFU")
plt.bar(x + width * 5, sca, width, alpha=1, color=colors[5], label="SCA")
plt.bar(x + width * 6, esca, width, alpha=1, color=colors[6], label="ESCA")
plt.xticks(np.arange(width * 3, 3 + width * 2 + 1, 1), labels=hardware)

plt.ylabel('Access Time')
plt.grid(True, axis='y')
plt.xlabel('Hardware')
plt.legend()
plt.show()
```

## Case 3. Criterions on TLB

```python
import numpy as np
import matplotlib.pyplot as plt


width = 0.15
x = np.arange(4)
fig, ax = plt.subplots(figsize=(15,6))
size = ['0', '128', '256', '512']

memory  = np.array([30836800, 27274300, 26706500, 25531200])
memory  = memory / (memory.max() + memory.min())

disk    = np.array([31200017160, 30677145420, 29773639080, 27515874900])
disk    = disk / (disk.max() + disk.min())

colors = ['#212121', '#323232', '#0D7377', '#3BACB6', '#82DBD8', '#B3E8E5']

plt.title("Hardware Access Time due to the TLB Size")
plt.plot(x, memory, color=colors[1], label='Memory')
plt.plot(x, disk, color=colors[2], label='Disk')

plt.xticks(np.arange(0, 4), labels=size)

plt.grid(True, axis='y')
plt.xlabel('TLB Size (Entry)')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```
